```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.worldometers.info/gdp/gdp-by-country/'
```

```python
page=requests.get(url)
```

```python
page
```

```python
soup=BeautifulSoup(page.text)
```

```python
soup.prettify
```

```python
gdp=soup.find(id='example2')
```

```python
gdp.prettify
```

```python
items=gdp.find("tbody")
```

```python
items
```

```python
rows=items.find_all('tr',style="")
```

```python
gdpinfo=[]
for row in rows:
    columns = row.find_all("td")
    gdp_info=[column.text.strip() for column in columns]
    gdpinfo.append(gdp_info)
```

```python
gdpinfo
```

```python
rank=[]
country=[]
gdpno=[]
gdpavr=[]
gdpgrow=[]
popu=[]
gdpperc=[]
share=[]
for i in gdpinfo:
    rank.append(i[0])
    country.append(i[1])
    gdpno.append(i[2])
    gdpavr.append(i[3])
    gdpgrow.append(i[4])
    popu.append(i[5])
    gdpperc.append(i[6])
    share.append(i[7])
```

```python
Data={'Rank':rank,'Country':country,'GDP (Nominal)':gdpno,'GDP (Abbrev)':gdpavr,'GDP
```

```python
import pandas as pd
```

```python
GdpInfo=pd.DataFrame(Data)
```

```python
GdpInfo.to_csv('Countries GDP.csv',index=False)
```
